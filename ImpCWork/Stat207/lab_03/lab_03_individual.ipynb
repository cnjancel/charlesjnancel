Create a boxplot for each of your numerical variables in this dataset.

### 2.3. Outlier Cleaning

1. If you see any outliers that are remarkably distinct from the rest of the other observations in any of these plots, drop them from the dataset.
2. How many new rows did you drop? 

```python
Q1 = flavor_cleaned[numerical_cols].quantile(0.25)
Q3 = flavor_cleaned[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier_conditions = (flavor_cleaned[numerical_cols] < (Q1 - 1.5 * IQR)) | (flavor_cleaned[numerical_cols] > (Q3 + 1.5 * IQR))

no_outlier_df = flavor_cleaned[~outlier_conditions.any(axis=1)]
```

```python
rows_dropped_after_outliers = flavor_cleaned.shape[0] - no_outlier_df.shape[0]
rows_dropped_after_outliers
```

### 2.5. Outlier Cleaning

1. Drop all significant outliers that you see in your figure above.
2. How many new rows did you drop?

```python
# Determine IQR for each numerical column pair and identify rows with two-variable outliers
outliers_2d_rows = set()

# Loop through each pair of numerical columns
for i in range(len(numerical_cols)):
    for j in range(i+1, len(numerical_cols)):
        col1 = numerical_cols[i]
        col2 = numerical_cols[j]
        
        # Calculate IQR for the two columns
        Q1_1, Q3_1 = no_outlier_df[col1].quantile([0.25, 0.75])
        Q1_2, Q3_2 = no_outlier_df[col2].quantile([0.25, 0.75])
        IQR_1 = Q3_1 - Q1_1
        IQR_2 = Q3_2 - Q1_2
        
        # Identify outliers for the two columns
        outlier_conditions_2d = (
            ((no_outlier_df[col1] < (Q1_1 - 1.5 * IQR_1)) | (no_outlier_df[col1] > (Q3_1 + 1.5 * IQR_1))) &
            ((no_outlier_df[col2] < (Q1_2 - 1.5 * IQR_2)) | (no_outlier_df[col2] > (Q3_2 + 1.5 * IQR_2)))
        )
        outliers_2d_rows.update(no_outlier_df[outlier_conditions_2d].index.tolist())
```

```python
cleaned_2d_df_IQR = no_outlier_df.drop(outliers_2d_rows)
```

```python
rows_dropped_2d_IQR = no_outlier_df.shape[0] - cleaned_2d_df_IQR.shape[0]
rows_dropped_2d_IQR
```

### 2.6. Outlier Cleaning Pros and Cons

1. When thinking about our research goal, what is one potential reason for why we may want to clean the outliers from our dataset?
2. When thinking about our research goal, what is one potential reason for why we may NOT want to clean the outliers from our dataset?

**Reason to Clean Outliers:**
- **Enhanced Model Generalization**: Removes undue influence on the regression line, leading to better predictions for the majority of data.

**Reason NOT to Clean Outliers:**
- **Potential Information Loss**: Outliers might represent unique, valuable cases. Removing them can limit insights and predictive capabilities for such instances.

## 4. Searching for Interaction Effects

1. Is there an interaction between how `Species` and `Aftertaste` effect `Flavor`? 
2. Is there an interaction between how `Species` and `Acidity` effect `Flavor`? 
3. Is there an interaction between how `Species` and `Body` effect `Flavor`? 
4. Is there an interaction between how `Color` and `Aftertaste` effect `Flavor`? 
5. Is there an interaction between how `Color` and `Acidity` effect `Flavor`? 
6. Is there an interaction between how `Color` and `Body` effect `Flavor`? 

You should create and interpret 6 visualizations to help you answer this.

```python
# Create plots to investigate interaction effects
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 18))

# List of numerical explanatory variables
variables = ['Data.Scores.Aftertaste', 'Data.Scores.Acidity', 'Data.Scores.Body']

# Create plots to investigate interaction effects using scatterplot and regplot

# 1. Interaction of Species with Aftertaste, Acidity, and Body on Flavor
for idx, var in enumerate(variables):
    sns.scatterplot(data=cleaned_2d_df_IQR, x=var, y='Data.Scores.Flavor', hue='Data.Type.Species', ax=axes[idx, 0], s=10)
    sns.regplot(data=cleaned_2d_df_IQR, x=var, y='Data.Scores.Flavor', scatter=False, ax=axes[idx, 0], line_kws={'linewidth':1})
    axes[idx, 0].set_title(f'Interaction of Species with {var.split(".")[-1]} on Flavor')

# 2. Interaction of Color with Aftertaste, Acidity, and Body on Flavor
for idx, var in enumerate(variables):
    sns.scatterplot(data=cleaned_2d_df_IQR, x=var, y='Data.Scores.Flavor', hue='Data.Color', ax=axes[idx, 1], s=10)
    sns.regplot(data=cleaned_2d_df_IQR, x=var, y='Data.Scores.Flavor', scatter=False, ax=axes[idx, 1], line_kws={'linewidth':1})
    axes[idx, 1].set_title(f'Interaction of Color with {var.split(".")[-1]} on Flavor')

plt.tight_layout()
plt.show()
```

**Left Column (Interaction of Species)**
1. **Aftertaste vs. Flavor**: Different slopes for the two species, indicating interaction.

2. **Acidity vs. Flavor**: Positive association for both species, but with varied slopes.

3. **Body vs. Flavor**: Positive association with a steeper slope for Robusta, suggesting interaction.

**Right Column (Interaction of Color)**

4. **Aftertaste vs. Flavor**: Varied relationships for different colors.
   
5. **Acidity vs. Flavor**: Some variations among colors, but less pronounced.

6. **Body vs. Flavor**: Slight variations in relationships among colors.

In summary, both `Species` and `Color` show interaction effects with the numerical variables on `Flavor`, affecting the relationships differently.
