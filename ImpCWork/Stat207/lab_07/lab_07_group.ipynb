### 2.2. Usability of an *Unscaled* Linear Regression Model

Suppose we were to just simply fit a linear regression model that predicts song `popularity` with your `first genre dataset` given the UNSCALED versions of our song metric explanatory variables. Would we be able to interpret the absolute value of the resulting slopes as an indication of how IMPORTANT the corresponding explanatory variables were when it comes to predicting `popularity` of songs from the `first genre dataset`? Explain.

No, we would not be able to interpret the absolute value of the resulting slopes as an indication of how important the corresponding explanatory variables were when it comes to predicting `popularity` of songs from the `first genre dataset` if we used the UNSCALED versions of our song metric explanatory variables. Here's why:

1. **Different Units and Ranges**: Each of the song metric explanatory variables has different units and ranges. For instance, `bpm` might range from 60 to 200, while `dB` might range only a few units. A small change in `dB` might have a significant impact on popularity, but the same absolute change in `bpm` might not.

2. **Magnitude of Coefficients**: In an unscaled linear regression model, the magnitude of the coefficients (slopes) is influenced by the scale of the explanatory variables. A variable with a larger scale (like `dur` which represents song duration in seconds) might have a smaller coefficient compared to a variable with a smaller scale. This doesn't mean the variable with the larger scale is less important; it's just a consequence of the different scales.

3. **Interpretation Difficulties**: Without scaling, a unit change in one variable might not be comparable to a unit change in another variable. For example, a one-unit change in `bpm` is not equivalent to a one-unit change in `dnce` in terms of their impact on the song's `popularity`.

### 2.3. z-Score Scaling the Explanatory Variables

Next, scale each of your explanatory variables in your `first genre dataset` using the z-score scaling method that we discussed in class.

```python
columns_to_scale = ['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch']
genre1_data = spotify_df[spotify_df['top genre'] == 'dance pop']

genre1_scaled = genre1_data.copy()
genre1_scaled[columns_to_scale] = (genre1_data[columns_to_scale] - genre1_data[columns_to_scale].mean()) / genre1_data[columns_to_scale].std()

genre1_scaled[columns_to_scale].head() 
```

### 2.4. Linear Regression Model with the Scaled Explanatory Variables

Next, fit a linear regression model with your `first genre dataset`, predicting `popularity` with each of your 9 *scaled song metric explanatory variables*.

```python
import statsmodels.api as sm

X_dance_pop = genre1_scaled[['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch']]
y_dance_pop = genre1_data['popularity']

X_dance_pop_const = sm.add_constant(X_dance_pop)

model_dance_pop = sm.OLS(y_dance_pop, X_dance_pop_const).fit()

model_dance_pop.summary()
```

### 2.5. Top 3 Most Important Song Metrics

### 2.6. Complicating Effect of Multicollinearity

Don't forget that when a linear regression model has issues with multicollinearity, our interpretations of the resulting slopes can be misleading and not reflect the relationships that exist between the explanatory variables and the response variable in our training dataset. This slope interpretation complication also holds for explanatory variables that have been scaled.

Does our linear regression model have an issue with multicollinearity? Explain.

To determine if our linear regression model has an issue with multicollinearity, we can look at a few indicators:

1. **Variance Inflation Factor (VIF)**: This is a common metric used to detect multicollinearity. If any variable has a VIF greater than 10 (some researchers use a threshold of 5), it's a sign that it's highly correlated with other predictors, indicating multicollinearity.
 
2. **Condition Number**: From the regression results, we see the "Cond. No." is 2.87. A condition number much greater than 30 is a sign of strong multicollinearity. In our case, the condition number is well below this threshold, suggesting that multicollinearity is not a significant issue.

3. **Coefficient Signs and Magnitudes**: If the signs or magnitudes of the coefficients don't align with what you'd expect based on domain knowledge or they change dramatically when a predictor is added/removed, it might suggest multicollinearity.

4. **Statistical Significance**: If several predictors are correlated, they might not be individually significant, but they could be significant when considered together.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = genre1_scaled[['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch']]
X = sm.add_constant(X)

vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data
```

Given the VIF values presented:

- A VIF of 1 indicates that there's no correlation between a given predictor and the other predictors.
- A VIF between 1 and 5 suggests that there's a moderate correlation, but it's typically not severe enough to warrant corrective measures.
- A VIF above 5 can be problematic, as it indicates potential multicollinearity.

Looking at the VIF values for your predictors:

- All predictors have VIF values well below 5.
- The highest VIF is for `nrgy` at $2.255317$, which suggests a moderate correlation with other predictors but is not alarmingly high.

Given these results, we can conclude that multicollinearity is not an issue in this linear regression model for the genre 1 songs in your dataset. The predictors are not highly intercorrelated.

```python
from scipy.stats import zscore
import statsmodels.formula.api as smf
# Filter out only the pop songs
genre2_songs = spotify_df[spotify_df['top genre'] == 'pop']

# Scaling the explanatory variables in genre2_songs dataset using z-score scaling
scaled_features_genre2 = genre2_songs[['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch']].apply(zscore)

# Adding the scaled features to the original genre2_songs DataFrame
scaled_genre2_songs = genre2_songs.copy()
scaled_genre2_songs[['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch']] = scaled_features_genre2

# Fitting a linear regression model to predict popularity using the scaled explanatory variables
model_genre2 = smf.ols("popularity ~ bpm + nrgy + dnce + dB + live + val + dur + acous + spch", data=scaled_genre2_songs).fit()
model_genre2.summary()
```
