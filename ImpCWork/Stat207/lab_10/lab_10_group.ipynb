**Host Information**
The dataset also contains the following information about the *host* of the given Airbnb listing:
* <u>host_is_superhost</u>: is the host a "superhost": t=True, f=False

## 1. Basic Data Examination and Manipulation

### 1.1. Basics
1. Read this csv file into a dataframe.

2. Display your dataframe.

3. Show how many rows your dataframe has.



```python
airbnb = pd.read_csv('seattle_airbnb_listings_cleaned.csv')
```

```python
airbnb_head = airbnb.head()
num_rows = len(airbnb)

airbnb_head, num_rows
```

### 1.2. 0/1 Response Variable

Create a new 0/1 variable in your dataset in which `superhosts=1` and `nonsuperhosts=0`. 

```python
airbnb['host_is_superhost'] = airbnb['host_is_superhost'].replace({'t':1, 'f':0})
airbnb.head()
```

### 1.3. Train-Test-Split

Create a training dataset and a test dataset from your dataframe using a random state of `101` and a test dataset size that is a random sample of about 20% of the observations from the original dataset.

```python
from sklearn.model_selection import train_test_split

# Define X and y

X = airbnb[['review_scores_rating', 'number_of_reviews', 'neighborhood', 'room_type', 'property_type']]
y = airbnb['host_is_superhost']

# One-Hot Encode Categorical Variables

X_encoded = pd.get_dummies(X, drop_first=True)

# Split the Data into Training and Testing Sets

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.20, random_state=101)
# Display the first few rows of the training data
X_train.head(), y_train.head()
```

## 2. Full Model and Test Log-Likelihood Function

In the interest of time, let's only consider using 5 potential explanatory variables in our logistic regression model, which will predict the `probability that a host is a superhost`:

* `review_scores_rating`
* `number_of_reviews`
* `neighborhood`
* `room_type`
* `property_type`

### 2.1. Fit the Model

Fit this "full" logistic regression model and show the summary output table. Use this summary output table to find the log-likelihood function value of this model and the training dataset.

```python
import statsmodels.api as sm
# Converting boolean values to integers
X_train_int = X_train.astype(int)

# Adding a constant to the features in the training set
X_train_const = sm.add_constant(X_train_int)

# Fitting the logistic regression model
log_reg_full = sm.Logit(y_train, X_train_const).fit()

# Showing the summary output
log_reg_full.summary()
```

### Interpretation of the Results:

1. **Log-Likelihood**: The log-likelihood of the model is -131.35, which gives us an idea of how well the model explains the data. The closer to 0, the better.
2. **Pseudo R-squared**: The value is 0.1879, indicating that around 18.79% of the variability in the response variable is explained by the model.
3. **Coefficients**: 
   - **const**: The intercept term is -25.5990, which is the log-odds of being a superhost when all other variables are 0.
   - **review_scores_rating**: For each additional point in review score rating, the log-odds of being a superhost increase by 0.2586.
   - **number_of_reviews**: For each additional review, the log-odds of being a superhost increase by 0.0148.
   - Other variables also have their respective coefficients, indicating their relationship with the response variable.
4. **P-values**: 
   - Variables such as `review_scores_rating`, `number_of_reviews`, and `neighborhood_Broadway` have p-values less than 0.05, suggesting that they are statistically significant predictors of being a superhost.

### 2.2. Training Data Predictive Probabilities

Calculate the predictive probabilities for all observations in the **training** dataset with model. **Call the pandas series that contains each of these predictive probabilities `phat_train`**.

*Note: In addition to using the **.predict()** function to predict the **predictive probability** of a single observation with a logistic regression model, we can use **.predict(YOUR_DATAFRAME)** to predict the predictive probabilities for all observations in a given dataframe with a logistic regression model.*

```python
# Calculating the predictive probabilities for the training data
phat_train = log_reg_full.predict(X_train_const)

# Showing the first few values of phat_train
phat_train.head()
```

### 2.3. Training Log-Likelihood Function Value

Assuming that your training dataframe was called `df_train`, the code below should calculate the the log-likelihood function value for your training dataset and this model. (If not, modify the code below to contain the actual name of your training dataframe).

This **log_loss()** should take take the following values:
* the actual 0/1 response variable values
* the predictive probabilities of a given model

Verify that this output here is the same as what you got in 2.1.

```python
from sklearn.metrics import log_loss

# Calculating the logistic loss (negative log-likelihood)
logistic_loss = log_loss(y_train, phat_train)

# Adjusting the sign to get the log-likelihood
log_likelihood = -logistic_loss * len(y_train)
log_likelihood
```

The log-likelihood function value for the training dataset and the logistic regression model is approximately $−131.35$.

This value is very close to the log-likelihood value obtained in the summary output of the logistic regression model in step 2.1, confirming the consistency of the results.

### 2.4. Test Log-Likelihood Function Value

Use the code in 2.2 and 2.3 to now calculate the log-likelihood function value of the **test dataset** and this model.

```python
X_test = X_test.astype(int)
# Adding a constant to the features in the test set
X_test_const = sm.add_constant(X_test)

# Calculating the predictive probabilities for the test dataset
phat_test = log_reg_full.predict(X_test_const)

# Showing the first few predictive probabilities
phat_test.head()
```

## 3. Backwards Elimination Algorithm

Ideally, we'd like to train (with our training dataset) and select the logistic regression model that will yield the HIGHEST **test log-likelihood function (LLF) value** (in our case it will be the LLF that is the least negative). Let's use a backwards elimination algorithm to try to help us find this model with the highest test LLF value.

*Note: Remember that the backwards elimination algorithm is just an algorithm "template". You can use this algorithm template to evaluate other types of models (like logistic regression) in attempt to find the model that has the highest metric value (like the highest test LLF value).*

```python
# Showing the summary output of the full logistic regression model
log_reg_full.summary()
```

```python
# Dropping the variable with the highest p-value
X_train_reduced = X_train_const.drop(columns='neighborhood_Minor')
X_test_reduced = X_test_const.drop(columns='neighborhood_Minor')

# Refitting the logistic regression model without 'neighborhood_Minor'
log_reg_reduced = sm.Logit(y_train, X_train_reduced).fit()

# Showing the summary output of the reduced logistic regression model
log_reg_reduced.summary()
```

```python
# Checking the data types of the reduced training dataset
X_train_reduced.dtypes
```

```python
# Dropping the variable with the highest p-value
X_train_reduced = X_train_reduced.drop(columns='room_type_Private room')
X_test_reduced = X_test_reduced.drop(columns='room_type_Private room')

# Refitting the logistic regression model without 'room_type_Private room'
log_reg_reduced_corrected = sm.Logit(y_train, X_train_reduced).fit()

# Showing the summary output of the further reduced logistic regression model
log_reg_reduced_corrected.summary()
```

```python
# Calculating the predictive probabilities for the test dataset using the further reduced model
phat_test_further_reduced = log_reg_reduced_corrected.predict(X_test_reduced)

# Calculating the test log-likelihood function value for the further reduced model
test_llf_further_reduced = log_loss(y_test, phat_test_further_reduced)
test_llf_further_reduced
```

### Comparison:

   * Test LLF of Previous Reduced Model: $0.5065$
   * Test LLF of Further Reduced Model: $0.4976$

Since the further reduced model has a lower (more negative) test log-likelihood function value than the previous reduced model, we will stop the backwards elimination algorithm here. The previous reduced model (before removing `room_type_Private room`) is the selected model.

This means that the model with the following variables is our final selected logistic regression model:

  *  `const`
  *  `review_scores_rating`
  *  `number_of_reviews`
  *  `neighborhood_Broadway`
  *  `neighborhood_Fremont`
  *  `neighborhood_Wallingford`
  *  `room_type_Shared room`
  *  `property_type_House`

This final model has a test log-likelihood function value of $0.5065$, indicating its performance on the test dataset.​​

## 4. Final Model

Write out your final logistic regression model in equation form using the three different representations that we discussed in class.
1. The one that predicts the **probability** the host is a superhost on the left hand side.
2. The one that predicts the **odds** the host is a superhost on the left hand side.
3. The one that predicts the **log-odds** the host is a superhost on the left hand side.

Make sure that you use the appropriate notation that we discussed in class.

The final model includes the following variables and their respective coefficients:
- Intercept ($ \beta_0 $): -25.6420
- review_scores_rating ($ \beta_1 $): 0.2592
- number_of_reviews ($ \beta_2 $): 0.0148
- neighborhood_Broadway ($ \beta_3 $): -1.1647
- neighborhood_Fremont ($ \beta_4 $): -0.5164
- neighborhood_Wallingford ($ \beta_5 $): -0.5754
- room_type_Private room ($ \beta_6 $): 0.3025
- room_type_Shared room ($ \beta_7 $): -1.2460
- property_type_House ($ \beta_8 $): -0.3330

### 1. Probability Representation:
The probability that a host is a superhost is given by:

$$
P(Y = 1) = \frac{1}{1 + \exp(-(\beta_0 + \beta_1 \cdot \text{{review\_scores\_rating}} + \beta_2 \cdot \text{{number\_of\_reviews}} + \beta_3 \cdot \text{{neighborhood\_Broadway}} + \beta_4 \cdot \text{{neighborhood\_Fremont}} + \beta_5 \cdot \text{{neighborhood\_Wallingford}} + \beta_6 \cdot \text{{room\_type\_Private room}} + \beta_7 \cdot \text{{room\_type\_Shared room}} + \beta_8 \cdot \text{{property\_type\_House}}))}
$$

### 2. Odds Representation:
The odds that a host is a superhost is given by:

$$
\frac{P(Y = 1)}{1 - P(Y = 1)} = \exp(\beta_0 + \beta_1 \cdot \text{{review\_scores\_rating}} + \beta_2 \cdot \text{{number\_of\_reviews}} + \beta_3 \cdot \text{{neighborhood\_Broadway}} + \beta_4 \cdot \text{{neighborhood\_Fremont}} + \beta_5 \cdot \text{{neighborhood\_Wallingford}} + \beta_6 \cdot \text{{room\_type\_Private room}} + \beta_7 \cdot \text{{room\_type\_Shared room}} + \beta_8 \cdot \text{{property\_type\_House}})
$$

### 3. Log-Odds Representation:
The log-odds that a host is a superhost is given by:

$$
\log\left(\frac{P(Y = 1)}{1 - P(Y = 1)}\right) = \beta_0 + \beta_1 \cdot \text{{review\_scores\_rating}} + \beta_2 \cdot \text{{number\_of\_reviews}} + \beta_3 \cdot \text{{neighborhood\_Broadway}} + \beta_4 \cdot \text{{neighborhood\_Fremont}} + \beta_5 \cdot \text{{neighborhood\_Wallingford}} + \beta_6 \cdot \text{{room\_type\_Private room}} + \beta_7 \cdot \text{{room\_type\_Shared room}} + \beta_8 \cdot \text{{property\_type\_House}}
$$

In these equations:
- $ P(Y = 1) $ represents the probability that the host is a superhost.
- The variables (e.g., review_scores_rating, number_of_reviews, etc.) represent the corresponding values for a particular Airbnb listing.
- The coefficients (e.g., $ \beta_0, \beta_1, \ldots $) represent the estimated coefficients from the logistic regression model.

These representations help in understanding how changes in the explanatory variables are associated with changes in the probability, odds, and log-odds of being a superhost.
