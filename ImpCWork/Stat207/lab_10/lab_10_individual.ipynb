## <u>Case Study</u>: Predicting Airbnb Superhost Status

Suppose that you have a spare room in your house (in Seattle) and you are considering advertising this room to guests on Airbnb. Airbnb superhosts are considered to be experienced hosts who provide a shining example for other hosts, and extraordinary experiences for their guests. Once a host reaches Superhost status, a superhost badge will automatically appear on their listing and profile to help guests identify them. You would like to assess your chances of being named a superhost with your property.

The following dataset is a sample (assume random) of available Airbnb listings in Seattle, WA. These listings were collected in January 2016, and filtered to just contain listings from the five most popular Seattlbe neighborhoods (for Airbnb listings) and just contain listings that are either in a house or apartment property.

This dataset contains the following variables.

**Listing Information**
The dataset contains the following information about the Airbnb *listing*:
* <u>price</u>: price of the listing (per night)(in US dollars)
* <u>review_scores_rating</u>: the average rating of the listing [0,100] (100 is the best)
* <u>number_of_reviews</u>: the number of reviews for the listing
* <u>security_deposit</u>: the security deposit required for the listing (in US dollars)
* <u>cleaning_fee</u>: the cleaning fee required for the listing (in US dollars)
* <u>neighborhood</u>: the neighborhood of Seattle the listing is located in
* <u>property_type</u>: is the listing in a 'House' or 'Apartment'
* <u>room_type</u>: is the listing a 'Entire home/apt', 'Private room', or 'Shared room'
* <u>accommodates</u>: how many guests will the listing accommodate
* <u>bathrooms</u>:how many bathrooms does the listing have
* <u>beds</u>: how many beds does the listing have

**Host Information**
The dataset also contains the following information about the *host* of the given Airbnb listing:
* <u>host_is_superhost</u>: is the host a "superhost": t=True, f=False
* <u>host_has_profile_pic</u>: does the host have a profile pic in their bio: t=True, f=False
* <u>host_response_time</u>: how fast will the host respond to requests (on average)

## Research Goals

The primary research goal that we will pursue in this lab will be to build a predictive model that effectively predicts the `probability a Seattle Airbnb host is a **superhost**` for *new Seattle Airbnb listings/hosts* given some combination of the `remaining explanatory variables`.

#### 1.3.3. Outlier Cleaning Pros and Cons

We intend to use this dataset to build a logistic regression model which will predict the `probability that a host is a superhost` using some combination of the `remaining variables` for *new Seattle Airbnb listings*.

1. Give one potential reason why we may want to drop outliers that you might see in this dataset for this particular analysis.
2. Give one potential reason why we may NOT want to drop outliers that you might see in this dataset for this particular analysis.

*Note: Given that outlier cleaning is subjective, for the purpose of a reproducible homework assignment we will NOT drop any outliers in this assignment.*

### 1. Reason to Drop Outliers:

#### 1.1. Improved Model Accuracy:
- **Impact on Model**: Outliers can have a significant impact on the results of a logistic regression model, potentially skewing the model's predictions and reducing its accuracy.
- **Specific to Superhost Prediction**: In the context of predicting whether a host is a Superhost, outliers could represent unusual listings that do not follow the typical patterns of other Airbnb listings. For example, a listing with an exceptionally high price or an unusually low review score might be an outlier.
- **Benefit of Dropping**: By removing these outliers, the logistic regression model may be able to better capture the underlying patterns in the data, leading to more accurate predictions for typical Airbnb listings in Seattle.

### 2. Reason to Keep Outliers:

#### 2.1. Capture Extreme Cases:
- **Real-World Scenarios**: Some outliers might represent real-world scenarios and extreme cases that are actually possible, even if they are rare.
- **Learning from Extremes**: The model can learn from these extreme cases and might be better prepared to make predictions in a wider variety of scenarios, including those that are less common.
- **Superhost Status**: For predicting Superhost status, there might be some listings that are truly exceptional in some way (e.g., extremely high or low prices, very few or very many reviews) but still have a Superhost. Keeping these outliers in the dataset ensures that the model is exposed to these scenarios during training.

### Conclusion:
The decision to drop or keep outliers depends on the specific goals of the analysis and the nature of the data. In this case, since the goal is to predict the probability that a host is a Superhost for new Seattle Airbnb listings, it might be beneficial to carefully evaluate whether each outlier represents a data entry error, a rare but legitimate scenario, or something else.

### 2.3. Preliminarily Weeding Out Explanatory Variables from a Model

#### 1. Do we know for sure that these two explanatory variables that we selected in 2.1 and 2.2 will overfit the model?

No, we do not know for sure that these two explanatory variables will overfit the model. The preliminary analysis conducted in sections 2.1 and 2.2 was aimed at identifying variables that seem to have weak or no association with the response variable, rather than identifying variables that would lead to overfitting.

Overfitting occurs when a model is too complex, capturing noise in the training data as if it were a real pattern. This often happens when there are too many parameters in the model relative to the number of observations, or when the model is overly complex. To determine if a variable will contribute to overfitting, we would need to consider the complexity of the entire model, including all of the variables and interactions, not just the individual variables in isolation.

#### 2. Is it possible that an explanatory variable that exhibits a strong association with our intended response variable may STILL overfit the model? If so, how/why might this happen?

Yes, it is possible. Even if an explanatory variable exhibits a strong association with the response variable, it may still contribute to overfitting the model, especially in the following scenarios:

1. **High Dimensionality**: If the dataset has a large number of explanatory variables compared to the number of observations, the model can become too complex, leading to overfitting. This is known as the "curse of dimensionality."

2. **Collinearity**: If the explanatory variable is highly correlated with other variables in the model, it can cause multicollinearity, which can lead to unstable coefficient estimates and make the model more sensitive to changes in the model or data.

3. **Outliers**: If the strong association is driven by outliers or anomalies in the data, the model may perform poorly on new data that does not contain these outliers.

4. **Non-linear Relationships**: If the relationship between the explanatory variable and the response variable is non-linear, and a linear model is used, the model may not capture the underlying pattern correctly, leading to overfitting.

5. **Interaction Effects**: If there are interaction effects between explanatory variables that are not accounted for in the model, it can lead to misleading conclusions about the association between individual variables and the response variable.

6. **Data Quality**: If the data is noisy or contains errors, a variable may appear to have a strong association with the response variable when it actually does not.

In summary, while identifying variables with strong associations is important, it is also crucial to consider the overall complexity of the model, the relationships between variables, and the quality of the data to prevent overfitting. Regularization techniques, cross-validation, and model selection criteria (such as AIC or BIC) are commonly used methods to help prevent overfitting and select an appropriate model.
