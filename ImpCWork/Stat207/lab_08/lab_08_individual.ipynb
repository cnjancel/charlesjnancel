We intend to build linear regression models that will predict the `Flavor` of *new coffee brands* using some of the following explanatory variables.

Using some combination of these explanatory variables, we'd like to build and select a linear regression model which we can infer might perform well when it comes to predicting the `Flavor` of *new coffee brands*. Thus, let's create a training and a test dataset from our cleaned dataframe from 1.3.

Use a random state of `437` to do this. Your training dataset should be 80% of observations from your cleaned dataframe.

```python
from sklearn.model_selection import train_test_split

X = selected_df_cleaned.drop(columns=['Data_Scores_Flavor'])

y = selected_df_cleaned['Data_Scores_Flavor']

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=437)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 1.5. Features Matrices and Target Arrays

Use your training dataset and test dataset to create the following.
* Training features matrix
* Training target array
* Test features matrix
* Test target array

```python
X_train_head = X_train.head()
y_train_head = y_train.head()
X_test_head = X_test.head()
y_test_head = y_test.head()

X_train_head, y_train_head, X_test_head, y_test_head
```

### 1.6. Scaling the Features Matrices

We'd like to be able to compare the magnitudes of our slopes in our linear regression models to try to infer the corresponding explanatory variable's importance when it comes to predicting coffee flavor. 

So let's z-score scale our training features matrix as well as our test features matrix.

#### 1.6.1. Scaling the Training Features Matrix

Z-score scale each of the values in your TRAINING features matrix using the corresponding TRAINING data column means and standard deviations.

```python
mean_train = X_train.mean()
std_train = X_train.std()

X_train_scaled = (X_train - mean_train) / std_train

X_train_scaled.head()
```

#### 1.6.2. Scaling the Test Features Matrix

Z-score scale each of the values in your TEST features matrix using the corresponding TRAINING data column means and standard deviations.

```python
X_test_scaled = (X_test - mean_train) / std_train

X_test_scaled.head()
```

### 1.7. Multicollinearity

Does your training dataset have an issue with multicollinearity?

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

vif_data = pd.DataFrame()
vif_data["Variable"] = X_train_scaled.columns
vif_data["VIF"] = [variance_inflation_factor(X_train_scaled.values, i) for i in range(X_train_scaled.shape[1])]

vif_data
```

- All VIF values are below the common threshold of 10, indicating no strong evidence of multicollinearity.
- The highest VIF value in your table is for `Data_Scores_Aftertaste` at 3.48, which is still below the commonly used threshold of 10 but indicates that this predictor has the highest linear relationship with the other predictors.

Regardless, the conclusion is consistent: there isn't a severe multicollinearity issue among the predictors in the training dataset based on the VIF metric.


## 2. Non-Regularized Linear Regression Model

### 2.1. Training the Model
First, let's fit a non-regularized linear regression model using our training features matrix and target array.

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)

coefficients = lin_reg.coef_
intercept = lin_reg.intercept_

coefficients, intercept
```

### 2.2. Testing the Model

Next, calculate the test R^2 of this model.

```python
test_r2 = lin_reg.score(X_test_scaled, y_test)

test_r2
```

## 3. LASSO Linear Regression Models


### 3.1. Parameter Tuning

Let's try to find the value of $\lambda$ in a LASSO linear regression (that we will train with our training dataset) that *maximizes* the test $R^2$ value.

1. In a for loop, iterate through a series of $\lambda$ values that go from [0,0.005, 0.01, 0.015, 0.02,..., 0.5].
2. For each value of $\lambda$ do the following.
    * Train a LASSO linear regression model with the training dataset and this given $\lambda$.
    * Calculate the test R^2 value for this LASSO linear regression model.
    
Your final result of this problem should include a *dataframe* that has two columns:
* the $\lambda$ value
* the test R^2 that correspond to this $\lambda$ value.

*Hint: See the tutorial above for code suggestions).*

```python
from sklearn.linear_model import Lasso

# List of lambda values
lambdas = [i*0.005 for i in range(101)]  

# Lists to store lambda values and their corresponding test R^2 values
lambda_values = []
r2_values = []

# Loop through each lambda value
for lam in lambdas:
    # Train a LASSO model
    lasso_model = Lasso(alpha=lam, max_iter=10000)
    lasso_model.fit(X_train_scaled, y_train)
    
    # Calculate the test R^2
    test_r2_lasso = lasso_model.score(X_test_scaled, y_test)
    
    # Append to the lists
    lambda_values.append(lam)
    r2_values.append(test_r2_lasso)

# Create a dataframe with lambda values and corresponding test R^2 values
results_df = pd.DataFrame({
    'Lambda': lambda_values,
    'Test R^2': r2_values
})

results_df.head()
```

### 3.2. Test R^2 and $\lambda$ Relationship

Plot the relationship between the $\lambda$ values and the Test R^2 in a **line plot**.

*Hint: Your code might look something like this.*

`plt.plot(df_output['lambda'].values, df_output['test_r2'].values)`

```python
import matplotlib.pyplot as plt

# Given values
lambda_values = [0.000, 0.005, 0.010, 0.015, 0.020]
test_r2_values = [0.825777, 0.827315, 0.828891, 0.829586, 0.82918]

# Plotting the relationship
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, test_r2_values, '-o', color='blue')
plt.xlabel('Lambda (Regularization Strength)')
plt.ylabel('Test R^2')
plt.title('Relationship between Lambda and Test R^2')
plt.grid(True)
plt.show()
```

### 3.3. Highest Test R^2

Which of our fitted LASSO linear regression models do we think might perform the best when predicting coffee flavor for *new datasets*? Explain.

```python
# Create a dataframe using the values
df_output = pd.DataFrame({
    'lambda': lambda_values,
    'test_r2': test_r2_values
})

# Find the index of the maximum test R^2 value
max_r2_index = df_output['test_r2'].idxmax()

# Retrieve the corresponding lambda value
best_lambda = df_output.iloc[max_r2_index]['lambda']
best_r2 = df_output.iloc[max_r2_index]['test_r2']

best_lambda, best_r2
```

#### 4.5.3 Interpretation

Which regularization model (LASSO vs. ridge regression) is it easier to interpret which explanatory variables bring enough predictive power to the model?

1. **LASSO Regression**:
   - LASSO regression performs both variable selection and regularization. It can shrink some coefficients to exactly zero, effectively excluding them from the model.
   - When a coefficient is shrunk to zero in LASSO regression, it indicates that the corresponding explanatory variable does not bring significant predictive power to the model, given the other variables.
   - This makes LASSO particularly straightforward for interpretation because you can directly see which variables the model deems important (non-zero coefficients) and which it does not (zero coefficients).

2. **Ridge Regression**:
   - Ridge regression performs regularization but does not typically zero out coefficients entirely. Instead, it shrinks the magnitude of the coefficients.
   - All explanatory variables remain in the model, but their coefficients are reduced in magnitude, making it harder to directly infer the importance of each variable.
   - The ratio of the Ridge regression slopes to the non-regularized linear regression slopes (as we computed in the previous step) can provide insights into the extent of shrinkage and, indirectly, the importance of the variable. However, this interpretation is less direct than in LASSO.

In summary, **LASSO regression** generally makes it easier to interpret which explanatory variables bring enough predictive power to the model because of its inherent feature selection capability. If a variable's coefficient is zero in a LASSO model, it's clear that the variable is not deemed important, given the presence of other variables in the model. On the other hand, with Ridge regression, while you get a sense of the relative importance of variables based on the magnitude of shrinkage, there's no direct exclusion of variables, making interpretation slightly more nuanced.
