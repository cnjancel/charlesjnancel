### 3.6. Modality of Moisture

1. What types of `Species` and `Colors` of coffee tend to be associated with high moisture coffees (roughly in the high moisture mode)?
2. What types of `Species` and `Colors` of coffee tend to be associated with low moisture coffees (roughly in the low moisture mode)?

* For the low moisture mode:
   - Most common `Species`: Arabica
   - Most common `Color`: Green
* For the high moisture mode:
   - Most common `Species`: Arabica
   - Most common `Color`: Green

Both the low and high moisture modes predominantly comprise the Arabica species and the Green color.

```python
kde = sns.kdeplot(coffee11_cleaned['Data_Scores_Moisture']).get_lines()[0].get_data()

x, y = kde

low_moisture_mode = x[y.argmax()]
high_moisture_mode = x[len(x) - y[::-1].argmax() - 1]

low_moisture_mode, high_moisture_mode

sorted_indices = y.argsort()
top_two_peaks = sorted_indices[-2:]

low_moisture_mode, high_moisture_mode = sorted(x[top_two_peaks])

low_moisture_mode, high_moisture_mode
```
