### 4. Select a categorical variable from this dataset and create a barplot that shows the *counts* for each level of this categorical variable. [1 pt]

```python
df_agg = df.groupby('category')['subscribers'].sum().reset_index()
df_agg_sorted = df_agg.sort_values(by='subscribers', ascending=False)
plt.figure(figsize=(14,10))
sns.barplot(x=df_agg_sorted.category.values, y=df_agg_sorted.subscribers.values, palette='viridis', errorbar=('ci', 0)).set(title='Subscribers per Category')
plt.xlabel('Category')
plt.ylabel('Number of Subscribers (Millions)')
plt.xticks(rotation = 90);
```
